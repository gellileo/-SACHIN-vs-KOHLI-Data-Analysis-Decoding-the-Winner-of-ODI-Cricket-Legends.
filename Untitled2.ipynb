```python
#creating separate data frames for sachin and kohli
#kohli =2008 to present(ODIs)
#sachin =1989 to 2012(ODIs)
kohli_df=df[df["year"]>=2008]
sachin_df = df[(df.year >= 1994) & (df.year <= 2004)]
```
